## 3.1. What are Spatial Composite Indicators?

A **composite indicator (CI)** is a mathematical aggregation of a set of individual indicators that measure multi-dimensional concepts but usually have no common units of measurement [(Nardo et al., 2005)](https://www.sciencedirect.com/topics/computer-science/composite-indicator). 
***Spatial* composite indicators** simplify complex, multi-dimensional data into concise performance metrics used in policy and decision-making. While useful for visualizing intricate systems, these indicators are inherently subjective, as their components are chosen by decision-makers [(Şalap-Ayça, 2025)](https://i-guide.io/i-guide-vco/rethinking-spatial-composite-indicators-with-the-lens-of-machine-learning/).  

Examples of composite indicators: 
- Air Quality Index
- Land Quality Index
- Literacy Rate
- GDP
- Environmental Benefits Index
- Social Vulnerability Index 



## 3.2 Example: Social Vulnerability Index

The **Social Vulnerability Index (SVI)** “is a place-based index designed to identify and quantify communities experiencing social vulnerability such as poverty, lack of access to transportation, and crowded housing" [(CDC, 2024)](https://www.atsdr.cdc.gov/place-health/php/svi/index.html).

The SVI “uses 16 variables to identify communities that may need support before, during, or after disasters. These variables are grouped into four major areas of social vulnerability, then combined into a single measure of overall social vulnerability” [(CDC, 2024)](https://www.atsdr.cdc.gov/place-health/php/svi/index.html).

The SVI needs multiple indices because social vulnerability is a multifaceted concept encompassing various factors that influence a community's ability to cope with and recover from hazards or adverse events. By combining multiple indices, the SVI provides a more comprehensive and nuanced understanding of social vulnerability, allowing for a more targeted and effective allocation of resources and interventions.


## 3.3 Characteristics of Composite Indicators 

So with composite indicators as a whole, 
> *The use of a single numeric index to assess [...] across different measures can be useful to inform decision making.*

Let's dive into each aspect:

<details>
<summary> <b>1) <i>single numeric index</i></b> </summary>
  
Spatial composite indicators are ***simple***. They provide a summary of the multi-dimensional complex system that is not so easy to visualize. The various components of the SVI, for example, can be very difficult to comprehend when treated as one.
</details>
<br>

<details>
  <summary> <b>2) <i>assess</i></b> </summary>
  
Spatial composite indicators are ***synthetic or performance based***. A vulnerability index doesn’t really exist, but it is made up in order to create a metric that represents complex phenomena not directly measurable nor uniquely definable.
</details>
<br>

<details>
  <summary> <b>3) <i>different measures</i></b> </summary>
  
Spatial composite indicators are ***pseudo spatial***. Even though some of them were derived from spatial data and represented spatially, they do not reflect any spatial relationship of the phenomena. 
</details>
<br>

<details>
  <summary> <b>4) <i>may be</i></b> </summary>
  
Spatial composite indicators are ***subjective*** in selection. What goes into a composite indicator depends on who the decision maker is. You can include as many indicators as needed. They are also subjective in that how each parameter is weighted can vary.
</details>

So what can be *different* within spatial composite indicators?
Well, they are: 
- spatially inherited
- scale-dependent
- policy relevant
- representative.
